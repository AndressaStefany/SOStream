```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sostream import SOStream
```

```python
data_df = pd.read_csv('s1.txt', delimiter='    ', header=None, engine='python')
# data_df.head()
```

```python
data_df.plot.scatter(x=0,y=1)

sostream_clustering = SOStream(alpha = 0, min_pts = 3, merge_threshold = 50000)

for r in data_df.values:
    sostream_clustering.process(r)

a = np.array([c.centroid for c in sostream_clustering.M[-1]])

plt.scatter(a[:,0], a[:,1], color='orange')

plt.show()
```
